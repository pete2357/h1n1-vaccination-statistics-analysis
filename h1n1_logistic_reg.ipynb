```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
x=df_vac.loc[:,df_vac.columns!='h1n1_vaccine']
y=df_vac.loc[:,df_vac.columns=='h1n1_vaccine']
```

```python
#solver liblinear
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression(solver='liblinear', random_state=0)
model1.fit(x,y)
model1.intercept_
model1.coef_
```

```python
#Predictions
```

```python
y_pred=model1.predict(x)
```

```python
#confusion matrix
```

```python
from sklearn import metrics
```

```python
cm=metrics.confusion_matrix(y,y_pred)
print(cm)
```

```python
#Accuracy Score-correct predictions/total number of data points
```

```python
model1.score(x,y)
```

```python
#ROC curve- receiver operating characteristic curve
#tpr
#fpr
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

```python
y_pred_prob=model1.predict_proba(x)
fpr, tpr, thresholds=roc_curve(df_vac['h1n1_vaccine'], y_pred_prob[:,1])
```

```python
roc_auc=auc(fpr, tpr) #Area under curve
print(roc_auc)
```

```python
#ROC Curve
plt.title=('ROC curve for h1n1 Vaccine Classifier')
plt.xlabel=('False positive Rate(1-Specificity)')
plt.ylabel=('True positive rate (Sensitivity)')
plt.plot([0,1], [0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.plot(fpr, tpr, label='AUC='+str(roc_auc))
plt.legend(loc=4) #location of label
plt.show()
```

```python
i = np.arange(len(tpr))
roc = pd.DataFrame({'fpr' : pd.Series(fpr, index=i),'tpr' : pd.Series(tpr, index = i), '1-fpr' : pd.Series(1-fpr, index = i), 'tf' : pd.Series(tpr - (1-fpr), index = i), 'thresholds' : pd.Series(thresholds, index = i)})
roc.iloc[(roc.tf-0).abs().argsort()[:1]]
```

```python
#ROC Curve
# Plot tpr vs 1-fpr
import pylab as pl
fig, ax = pl.subplots()
pl.plot(roc['tpr'])
pl.plot(roc['1-fpr'], color = 'red')
pl.xlabel=('1-False Positive Rate')
pl.ylabel=('True Positive Rate')
pl.title=('ROC Curve')
ax.set_xticklabels([])
```
